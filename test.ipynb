```python
def minDeletions(arr):
    n = len(arr)
    
    # Edge case: if array has 0 or 1 element, it's already sorted or almost sorted
    if n <= 1:
        return 0

    # dp array to store the length of LIS ending at each index
    dp = [1] * n  # Each element is itself an increasing subsequence of length 1
    
    # Compute the LIS using dynamic programming
    for i in range(1, n):
        for j in range(i):
            if arr[i] > arr[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    
    # Length of the longest increasing subsequence
    lis_length = max(dp)
    
    # The minimum number of deletions is the total length minus the length of the LIS
    return n - lis_length

# Exemple d'utilisation :
arr = [3, 4, 2, 5, 1]
print(minDeletions(arr))  # RÃ©sultat : 1
```
